### STEP 2: WRANGLE DATA AND COMBINE INTO A SINGLE FILE

```python
# concatenate all 12 months of data into 1 dataframe df
df = pd.concat([aug20,sep20,oct20,nov20,dec20,jan21,feb21,mar21,apr21,may21,jun21,jul21], axis=0, ignore_index=True, sort=False)
df.head()
```

```python
df.tail()
```

```python
# check shape of dataset
df.shape
```

```python
# rename columns
df = df.rename(columns={'ride_id' : 'trip_id',
                        'rideable_type' : 'bikeid',
                        'started_at' : 'start_time',
                        'ended_at' : 'end_time',
                        'start_station_name' : 'from_station_name',
                        'start_station_id' : 'from_station_id',
                        'end_station_name' : 'to_station_name',
                        'end_station_id' : 'to_station_id',
                        'member_casual' : 'usertype'
                       })
df.head()
```

```python
# inspect df
df.info()
```

```python
# inspect unique values in each variable
for column in list(df.columns):
    print(f'{column}: {df[column].unique()}\n')
```

### STEP 3: CLEAN UP AND ADD DATA TO PREPARE FOR ANALYSIS

```python
# convert to datetime type
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

df.info()
```

```python
# calculate duration between start_time and end_time
df['ride_length'] = (df['end_time'] - df['start_time'])

# convert ride_length to minutes and hours
df['ride_length_min'] = df['ride_length']/np.timedelta64(1,'m')
df['ride_length_hr'] = df['ride_length']/np.timedelta64(1,'h')

df.head()
```

```python
# create day of week feature
df['day_of_week'] = df['start_time'].dt.dayofweek  # Mon-0, Sun-6 

df['day_of_week'] = df['day_of_week'].replace({0:'Mon',
                                               1:'Tue',
                                               2:'Wed',
                                               3:'Thu',
                                               4:'Fri',
                                               5:'Sat',
                                               6:'Sun'
                                              }
                                             )
# convert datatype
# df['day_of_week'] = df['day_of_week'].astype(str)

# check unique values 
df['day_of_week'].unique()
```

```python
# create month of trip feature
df['month'] = df['start_time'].dt.month

df['month'] = df['month'].replace({1:'Jan',
                                   2:'Feb',
                                   3:'Mar',
                                   4:'Apr',
                                   5:'May',
                                   6:'Jun',
                                   7:'Jul',
                                   8:'Aug',
                                   9:'Sep',
                                   10:'Oct',
                                   11:'Nov',
                                   12:'Dec'
                                  }
                                 )                                             

# check unique values 
df['month'].unique()
```

```python
# create time of trip feature
df['time'] = df['start_time'].dt.time
df['time'].head()
```

```python
# create hour of trip feature
df['hour'] = df['start_time'].dt.hour
df['hour'].unique()
```

```python
df.head()
```

```python
# histogram of 'ride_length_min'
df['ride_length_min'].hist(bins=5)
plt.show()
```

```python
# check number of negative or zero values
(df['ride_length_min'] <= 0).sum()
```

```python
# check percentage of negative or zero values
(df['ride_length_min'] <= 0).mean()
```

```python
# remove rows with negative or zero values
df = df[df['ride_length_min'] > 0]

df.shape
```
